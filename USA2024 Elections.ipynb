- Allan Lichtman Keys thanks to BeautifulSoup üç≤  (Scrape \& Creation of the Lichtman dataframe)

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/The_Keys_to_the_White_House"
r = requests.get(url)
soup = BeautifulSoup(r.text ,"html.parser")
```

```python
tables = pd.read_html(str(soup))
tables
```

```python
for index, table in enumerate(tables): #first lets select all the tables by indexes just to check 
    print(f"Table{index}:")
    print(table.head())
```

```python
real_table_index = 1
real_table = tables[real_table_index]
real_table
```

```python
real_table.drop([2,3], inplace=True) #drop two that have different format
```

```python
real_table.reset_index(drop=True, inplace=True)
real_table #reset the indexes from the drop
```

```python
real_table["Election"] = real_table["Election"].astype(int)
bs1 = real_table[real_table['Election'] >= 1976]
bs1 #bs1 has data from one table of the tables that we wanted
```

```python
real_table_index2 = 2
real_table2 = tables[real_table_index2]
real_table2
```

```python
bs2 = real_table2
```

```python
bs_combined = pd.concat([bs1, bs2], axis = 0)
bs_combined = bs_combined.reset_index(drop=True)
bs_combined
```

```python
bs_combined_1 = bs_combined.drop("Presumed winner", axis=1)#dropar a coluna presumed winner almost no info
bs_combined_1
```
