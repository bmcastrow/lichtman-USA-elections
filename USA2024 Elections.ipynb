```python
url = "https://en.wikipedia.org/wiki/The_Keys_to_the_White_House"
r = requests.get(url)
soup = BeautifulSoup(r.text ,"html.parser")
```

```python
tables = pd.read_html(str(soup))
tables
```

```python
for index, table in enumerate(tables): #first lets select all the tables by indexes just to check 
    print(f"Table{index}:")
    print(table.head())
```

```python

df_combination = df_final_1.merge(bs_combined_1, on='year', how='left')
df_combination
```

```python
df_combination_drop1 = df_combination.drop("Election", axis=1)
df_combination_drop1
```

```python
#lets replace: make those True and False clean
df_combination_drop1 = df_combination_drop1.replace(r'\[.*\]', '', regex=True)
df_combination_drop1
```

```python
null_values_check = df_combination_drop1["Predicted winner"].isnull() #ver os nan values na coluna predicted winner do datafframe df combination drop
null_rows = df_combination_drop1[null_values_check][["year", "Predicted winner"]] #essesncialmente vamos ao nosso dfcombinationdrop e aplicamos a condi√ßao da variavel anterior de so mostrar nan values da predicted winner e depois sim damos display tanto dos predicted winners como dos years
null_rows #here we check which years dont have the predicted winner 
#*explanation for ze*
```

```python
df_combination_drop1
```

```python
df_combination_drop1["Predicted winner"] = df_combination_drop1["Predicted winner"].fillna("NO PREDICTION")
df_combination_drop1["vote_percentage"] = df_combination_drop1["vote_percentage"].round(2)
df_combination_drop1

```

```python
df_combination_drop1["candidate"] = df_combination_drop1["candidate"].str.split(', ').map(lambda x : ' '.join(x[::-1]))
df_combination_drop1
```

```python
df_combination_drop1["party_simplified"] = df_combination_drop1["party_simplified"].replace("DEMOCRAT", "DEMOCRATIC")
```

```python
df_combination_drop1.dtypes
```

```python
bool_columns = [
    'Party mandate',
    'No primary contest',
    'No third party',
    'Strong short-term economy',
    'Major policy change',
    'No social unrest',
    'No scandal',
    'No foreign or military failure',
    'Major foreign or military success',
    'Charismatic incumbent',
    'Uncharismatic challenger'
]

for col in bool_columns:
    df_combination_drop1[col] = df_combination_drop1[col].map({'True': True, 'False': False}).astype(bool)
```
